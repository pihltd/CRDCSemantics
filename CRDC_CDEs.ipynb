```python
import pandas as pd
import requests
import pprint
```

```python
devcadsrurl = "https://cadsrapi-dev.cancer.gov/rad/NCIAPI/1.0/api/DataElement/getCRDCList"
prodcadsrurl = "https://cadsrapi.cancer.gov/rad/NCIAPI/1.0/api/DataElement/getCRDCList"
```

```python
jsonresults = requests.get(prodcadsrurl, headers = {"accept":"application/json"})
pprint.pprint(jsonresults)
```

```python
jsonlist = jsonresults.json()['CRDCDataElements']    
```

```python
df = pd.DataFrame()
for cde in jsonlist:
    df = pd.concat([df, pd.DataFrame([cde])], ignore_index = True)
```

```python
df.head()
```

The values that DH cares about are:
- CDE Public ID
- CRDC Name
- VD Type
- CRDC Defintion
- Workflow Status
- Owned By
- Deep Link
- permissibleValues
- -- Permissible Value, Concept code

```python
print(df[['CDE Public ID', 'CRDC Name', 'Owned By', 'permissibleValues','Coding Instruction']])
```

```python
#pprint.pprint(jsonlist)
for cde in jsonlist:
    if cde['VD Type'] == 'Enumerated':
        print(("%s\t%s\t%s\n") % (cde['CRDC Name'], cde['CDE Public ID'], cde['Coding Instruction']))
        if cde['permissibleValues'] is not None:
            #print(("ID:\t%s\tName:\t%s\tValues:\t") % (cde['CDE Public ID'], cde['CRDC Name']))
            for pv in cde['permissibleValues']:
                print(("%s\t%s\n") % (pv['Permissible Value'], pv['Concept Code']))
    print("\n")
```
